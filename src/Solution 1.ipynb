```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values from 0 to 10
x = np.arange(0, 11, 1)
y = x ** 2  # Compute y values as the square of x

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o', linestyle='-', color='b', label='y = xÂ²')

# Customize the plot
plt.title('Square Function Plot')
plt.xlabel('X values')
plt.ylabel('Y values (X squared)')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
